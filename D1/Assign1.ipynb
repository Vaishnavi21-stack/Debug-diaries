```python
# Given string
text = "Hello Vaishnavi, welcome to Python programming"

# a. First character
print("First character:", text[0])

# b. Last character
print("Last character:", text[-1])

# c. First 5 characters
print("First 5 characters:", text[:5])

# d. Last 11 characters
print("Last 11 characters:", text[-11:])

# e. Reverse the string
print("Reversed string:", text[::-1])

# f. Extract the word "Python" (safe method)
start = text.find("Python")
end = start + len("Python")
print("Word 'Python':", text[start:end])
```

```python
# Given string
strM = "Python beginner tutorial"

# a. Convert to uppercase
print("Uppercase:", strM.upper())

# b. Convert to lowercase
print("Lowercase:", strM.lower())

# c. Capitalize (first letter uppercase, rest lowercase)
print("Capitalized:", strM.capitalize())

# d. Count occurrences of 't'
print("Count of 't':", strM.count('t'))

# e. Replace "Python" with "Machine Learning"
new_str = strM.replace("Python", "Machine Learning")
print("Replaced string:", new_str)
```

```python
# Create tuples
t1 = (10, 20, 30)
t2 = (40, 50, 60)

# a. Concatenate tuples
t_combine = t1 + t2
print("Concatenated tuple:", t_combine)

# b. Repeat elements 3 times
print("Repeated tuple:", t_combine * 3)

# c. Access 3rd element (index starts from 0)
print("3rd element:", t_combine[2])

# d. First three elements
print("First three elements:", t_combine[:3])

# e. Last three elements
print("Last three elements:", t_combine[-3:])
```
