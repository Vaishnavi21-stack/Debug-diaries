```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
def check_even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

number = int(input("Enter a number: "))
print(check_even_odd(number))
```

```python
#arguments*args
def greetings(greet,*name):
  for n in name:
    print(greet,n)
```

```python
greetings("hi",'vaish','chinni','vaishnavi')
```

```python
# **kargs = key arguments
def keyargs(**name):
    print("hi", name['fname'], name['lname'], name['fullname'])

keyargs(fname='vaishnavi', lname='lolabhattu', fullname='vaishnavi lolabhattu')
```
